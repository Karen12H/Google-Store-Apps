# Google Apps 商店的数据分析

## 简介

相信大家对移动应用商店都不陌生吧。Google Play Store（Google Play商店）是谷歌官方的软件应用商店，拥有上架软件数十万款，下载量更是突破了20亿次，为了手机用户提供了极为广泛的应用选择，很受大家的欢迎。



本数据集(googleplaystore.csv)包含了 Google Play 商店中 App 的数据。该数据是Kaggle中 [Google Play Store Apps](https://www.kaggle.com/lava18/google-play-store-apps) 的一部分，其中包含 Google Play 商店中 10k+ 应用软件的信息。

数据中的变量含义解释：
```
App: 应用的名称，字符变量。
Category: 应用所属的分类，字符变量。
Rating: 某应用的用户评分，数值变量。
Reviews: 某应用获得的用户评论数量，数值变量。
Size: 某应用的所占存储空间的大小，字符变量。
Installs: 用户安装和下载某应用的次数，字符变量。
Type: 付费或免费，分类变量。
Price: 价格，字符变量。
Content Rating: 应用商店针对内容给出的年龄评级组 - Children / Mature 21+ / Adult，分类变量。
Genres: 类型/流派，一个应用可以属于多个流派，比如音乐、游戏、家庭等，字符变量。
Last Updated: 应用最新更新的日期，字符变量。
Current Ver: 当前应用的版本，字符变量。
Android Ver: 安装该应用所需要的最低安卓版本，字符变量。

```

````python
# TO DO: load pacakges
import pandas as pd
import numpy as np
import seaborn as sns
% matplotlib inline
````

````python
# TO DO: load the dataset
df = pd.read_csv("googleplaystore.csv")
df
````

````python
# TO DO: check the dataset general info
df.info()
````

````python
# TO DO: clean the data (optional: only there are problems)

print(df['App'].unique().size)
````

````python
print(df['Category'].value_counts(dropna=False))
print(df[df['Category']=='1.9'])
````

````python
# drop error
df.drop(index=10472,inplace=True)
````

````python
print(df['Rating'].value_counts(dropna=False))
````

````python
df['Rating'].fillna(value=df['Rating'].mean(),inplace=True)
````

````python
print(df['Rating'].value_counts(dropna=False))
print(df['Reviews'].str.isnumeric().sum())
````

````python
df['Reviews'] = df['Reviews'].astype('i8')
df.describe()
````

````python
df['Size'].value_counts()
````

````python
df['Size']=df['Size'].str.replace('M','e+6')
df['Size']=df['Size'].str.replace('k','e+3')
df['Size']=df['Size'].str.replace('Varies with device', '0')
df['Size']=df['Size'].astype('f8')
df['Size'].replace(0, df['Size'].mean(),inplace=True)
df['Size']
````

````python
df.describe()
````

````python
df['Installs'].value_counts()
````

````python
df['Installs']=df['Installs'].str.replace('+', '')
df['Installs']=df['Installs'].str.replace(',', '')
df['Installs']=df['Installs'].astype('int')
print(df.describe())
````

````python
print(df['Type'].value_counts(dropna=False))
print(df[df['Type'].isnull()])
````

````python
df.drop(index=9148, inplace=True)
````

````python
df.drop_duplicates('App', inplace=True)
````

````python
df.describe()
````

## 数据探索分析

````python
# In exploratory data analysis, please make sure of using statistics and visualizations
df.Category.unique().size
````

````python
Category_App = df.groupby('Category').count().sort_values('App', ascending = False)
Category_App
````

````python
plt.figure(figsize=(30,10),dpi=40)
Category_App.plot(kind='barh')
plt.xlabel('numbers of Apps')
plt.title('Distribution of Apps')
plt.show()
````

````python
top_10 = df.groupby('Category').count()['App'].sort_values(ascending=False)[:10]
print(top_10)
````

````python
plt.figure(figsize=(20,10),dpi=80)
x = top_10.index
y = top_10.values

for a, b in zip(x, y):
    plt.text(a,b,b, ha='center',va='bottom', fontsize=30)
plt.bar(x, y, width= 0.5)
plt.xlabel('Category of Apps')
plt.ylabel('number of Apps')
plt.title('Top 10 Apps Category')
plt.show()
````

````python
Ca_Rating_mean = df.groupby('Category').mean()['Rating'].sort_values(ascending=False)
Ca_Rating_mean
````

````python
df.groupby('Type')['App'].count()
````

````python
df.groupby('Type').sum()['Installs'].sort_values(ascending=False)
````

````python
Ty_Cat = df.groupby(['Type', 'Category']).mean()
(Ty_Cat['Reviews']/Ty_Cat['Installs']).sort_values(ascending=False)
````

## 得出结论

**问题**：下载量最高的 Apps 有什么特征？

**答案**：Family 类的 Apps 拥有1831的下载量，是Google商店里下载量最高的 Apps类型。然而Family类的Apps评分却不是最高，只有4.18。 评分最高的Apps却是共有64个Apps的Event类。根据分析而得出结论，Family类由于有最多的Apps数量进而提高了下载量。因为下载量和Apps数量的相比其他类型Apps多，评分也多。高低不一的评分平均下来影响到最终评分，所以Family类的Apps最终评分没有Apps数量比较少的Event类Apps高。收费的Apps比免费的Apps评论比率高。综合以上分析，评论漱和安装数量相关比率为0.625。
